```python
import tkinter as tk
```

##### **Creating a window**
---
1. .Tk()
2. .title()
3. .geometry()

4. .mainloop()

```python
root = tk.Tk()
root.title("To-Do List")
root.geometry("400x400")

# root.mainloop()
```

##### **Basic Widgets**
---
1. Buttons: .Button(text="",command=func_nameOnly,bg="",fg="",font=("fontType",size,"bold/italics"),width=,height=,activebackground="",activeforeground=""
state=DISABLED/ENABLED,relief=RAISED/FLAT/SUNKEN/GROOVE/RIDGE,bd=,cursor="")

```python
btn = tk.Button(text="Button1",command= lambda: print("Button1 clicked"), bg="blue", fg="white",font=("Arial", 12, "bold"),width=10, height=2,activebackground="lightblue", activeforeground="white",relief=tk.RAISED, bd=5,cursor="hand2")

btn.pack(pady=20)
```

2. Labels: .Label(windowObject, text="",bg="",fg="",font=(),padx=,pady)

```python
tk.Label(root, text="Hello, Tkinter!",fg="white",bg="black").pack()
```

3. Entry(windowObject,width="",bg="",fg="",font=(),show="")

```python
entry = tk.Entry(root,show="*")
entry.pack()
entry.insert(0, "Type something...")
entry.bind("<Return>", lambda event: print("Entered:", entry.get()))
```

4. Text(windowObject,height=,width=,bg="",fg="")

```python
text = tk.Text(root, height=5, width=30)
text.pack()
text.insert(tk.END, "This is a Text widget.\nYou can write multiple lines here.")
text.bind("<Control-s>", lambda event: print("Ctrl+S pressed. Text content:", text.get("1.0", tk.END)))
root.mainloop()
```
