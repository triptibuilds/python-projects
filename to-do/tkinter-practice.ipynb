`.Buttons()`: .Button(text="", command=func_nameOnly,bg="", fg="", font=("fontType",size,"bold/italics"), width=, height=, activebackground="",activeforeground="", state=DISABLED/ENABLED, relief=RAISED/FLAT/SUNKEN/GROOVE/RIDGE, bd=, cursor="")

3. Entry(windowObject,width="",bg="",fg="",font=(),show="")

```python
entry = tk.Entry(root,show="*")
entry.pack()
entry.insert(0, "Type something...")
entry.bind("<Return>", lambda event: print("Entered:", entry.get()))
```

4. Text(windowObject,height=,width=,bg="",fg="")

```python
text = tk.Text(root, height=5, width=30)
text.pack()
text.insert(tk.END, "This is a Text widget.\nYou can write multiple lines here.")
text.bind("<Control-s>", lambda event: print("Ctrl+S pressed. Text content:", text.get("1.0", tk.END)))
```

`.Checkbutton()`: .Checkbutton(root, text="", command="", variable=IntVar/BooleanVar, offvalue=, bg=, fg=, font=(), state=, height=, width=,
anchor=, )
controlled via *.get()* and *.set()*

```python
cb_var = tk.BooleanVar()
cb = tk.Checkbutton(root, text="Agreed", variable=cb_var, command = lambda : print("Checkbox:",cb_var.get()), bg="lightgray", fg="black", font=("Arial", 10), activebackground="gray", activeforeground="white", relief=tk.FLAT, bd=0, cursor="hand2", onvalue=True, offvalue=False)
cb.pack(pady=10)
```

`.Listbox()`: .Listbox(root, height=, width=, selectmode= "single/multiple/browse/extended", exportselection=, )
|METHODS|
|-----------------------|
|*lb.insert(tk.END,[])*|
|*lb.get(lb.curselection())*|
|*lb.delete(0)*|
|*lb.delete(0,tk.END)*|
|*lb.get(0,tk.END)*|


```python
lb = tk.Listbox(root, selectmode=tk.BROWSE,exportselection=0)
lb.pack()
for item in ["Item 1", "Item 2", "Item 3", "Item 4"]:
    lb.insert(tk.END, item)

bttn = tk.Button(root, text="SHOW SELECTED", command= lambda : print("seleceted items:",[lb.get(i) for i in lb.curselection()]), bg="green", fg="white", font=("Arial", 10, "bold"), width=15, height=2, activebackground="lightgreen", activeforeground="white", relief=tk.RAISED, bd=5, cursor="hand2")
bttn.pack()

```

```python
# Practicing grid layout
# parameters: row, column, rowspan, columnspan, sticky, padx, pady, ipadx, ipady
# sticky: n, s, e, w, ne, nw, se, sw
# padx, pady: external padding (space outside the widget)
# ipadx, ipady: internal padding (space inside the widget)
# grid.rowconfigure(row, weight), grid.columnconfigure(column, weight)
# weight: controls how extra space is distributed among rows and columns when the window is resized. Higher weight means more space will be allocated to that row or column.


gWin = tk.Tk()
gWin.configure(bg="lightblue")
gWin.title("Practicing grid")
gWin.geometry("600x550")

# buttons, labels, checkbox, radiobutton, listbox, entry, text, frame
for i in range(3):
    gWin.grid_columnconfigure(i, weight=1)
gWin.grid_rowconfigure(0, weight=1)


# creating frames for better organization of widgets
frame1 = tk.Frame(gWin, bg="lightblue",bd=5)
frame2 = tk.Frame(gWin, bg="lightgray",bd=5,relief=tk.RIDGE)
frame3 = tk.Frame(gWin, bg="lightblue",bd=5)


# placing frames in the grid
frame1.grid(row=0,column=0,sticky='nsew')

frame2.grid(row=0,column=1,sticky='nsew')
frame2.grid_rowconfigure(0, weight=1)
frame2.grid_rowconfigure(1, weight=5)
frame2.grid_rowconfigure(2, weight=1)
frame2.grid_columnconfigure(0, weight=3)
frame2.grid_columnconfigure(1, weight=1)

frame3.grid(row=0,column=2,sticky='nsew')

# adding widgets to frame2
toDoLabel = tk.Label(frame2, text="TASKS LIST",bg="lightgray",fg="black",font=("Arial", 16, "bold"))
toDoLabel.grid(row=0,column=0,columnspan=2,sticky='ew')

taskList = tk.Listbox(frame2,bg='white',fg="black")
taskList.grid(row=1,column=0,columnspan=2,sticky='nsew',padx=10,pady=10)


entry = tk.Entry(frame2)
entry.grid(row=2,column=0,padx=10,pady=10,sticky='ew',ipady=5)

# button action to add task to listbox
def add_task():
    task = entry.get()
    if task:
        taskList.insert(tk.END,task)
        entry.delete(0,tk.END)

addBtn = tk.Button(frame2,text="ADD TASK",bg="blue",fg="white",command=add_task)
addBtn.grid(row=2,column=1,padx=10,pady=10,ipadx=5,ipady=2)



gWin.mainloop()
```
