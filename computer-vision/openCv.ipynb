```python
import numpy as np

kernel = np.ones((3,3),np.uint8)

img = cv.imread("download.jpg")

imgGray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
imgBlur = cv.GaussianBlur(img,(5,5),0)
imgCanny = cv.Canny(imgBlur,70,140)
imgDilation = cv.dilate(imgCanny,kernel,iterations=2)
imgErode = cv.erode(imgDilation,kernel,iterations=1)

cv.imshow("Gray",imgGray)
cv.imshow("Blur",imgBlur)
cv.imshow("Canny",imgCanny)
cv.imshow("Dilation",imgDilation)
cv.imshow("Erosion",imgErode)
cv.waitKey()
cv.destroyAllWindows()
```

```python
img = cv.imread("download.jpg")
h, w, c = img.shape

new_w = 400
new_h = int((new_w/w) * h)

resizeD = cv.resize(img,(new_w,new_h))
resizeF = cv.resize(img,None,fx=2,fy=2)

cv.imshow("ResizeD",resizeD)
cv.imshow("ResizeF",resizeF)
cv.waitKey()
cv.destroyAllWindows()
```

```python
img = cv.imread("download.jpg")

cropped = img[50:100,50:100]

cv.imshow("Normal",img)
cv.imshow("Cropped",cropped)

cv.waitKey()
cv.destroyAllWindows()
```

```python
img = cv.resize(img,None,fx=2,fy=2)
h, w, c = img.shape
area = 100

x = h//2
y = w//2

cropped = img[x-area:x+area,y-area:y+area]
cv.imshow("Mid-part",cropped)
cv.waitKey()
cv.destroyAllWindows()
```

```python
roi = cv.selectROI("Select",img)
x, y, w, h = roi

cropped = img[y:y+h,x:x+h]
cv.imshow("Mid-part",cropped)
cv.waitKey()
cv.destroyAllWindows()
```

```python
import numpy as np

mat = np.zeros((512,512,3))
# mat[200:300,200:300] = 0,0,255

cv.line(mat,(0,0),(mat.shape[1],mat.shape[0]),(255,255,255),3)
cv.rectangle(mat,(0,0),(300,300),(0,0,255),2) #cv.FILLED
cv.circle(mat,(400,50),40,(255,255,0),3)
cv.putText(mat," opencv ",(100,400),cv.FONT_HERSHEY_DUPLEX,1.5,(150,0,0),1)

cv.imshow("matrix",mat)
cv.waitKey()    
cv.destroyAllWindows()
```

```python
img = cv.imread("cards.jpg")

width,height = 250,350
pts1 = np.float32([[111,219],[287,188],[154,482],[352,440]])
pts2 = np.float32([[0,0],[width,0],[0,height],[width,height]])

matrix = cv.getPerspectiveTransform(pts1,pts2)
imgOutput = cv.warpPerspective(img,matrix,(width,height))

cv.imshow("Cards",imgOutput)
cv.waitKey()
cv.destroyAllWindows()
```
